```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
def sigma_pred(x, a, b):
    return (a / np.power(x, 2)) + (b / np.power(x, 3))
```

```python
sigma = np.array([0.372065,0.274510,0.194971,0.132350,0.092963, 0.069424, 0.052824])
end = sigma.shape[0] + 1
k_lin = np.linspace(start=1, stop=end, num=100)
k = np.arange(1, end)
pars, cov = curve_fit(f=sigma_pred, xdata=k[2::], ydata=sigma[2::], p0=[0, 0], bounds=(-np.inf, np.inf))
sig_pred = sigma_pred(k_lin, *pars)
```

```python
plt.figure(figsize=(8,6))
plt.plot(k, sigma, linestyle='none', marker='x', color='blue')
plt.plot(k_lin, sig_pred, linestyle='dashed', color='red')
plt.xlabel(r'$k$')
plt.ylabel(r'$\sigma_k$')
plt.ylim(0, 0.03 + np.max(sigma))
plt.legend([r'$\sigma_k$', r'$\sigma_k = %.3f / k^2 %+.3f / k^3$' % tuple(pars) ],
           prop={'size' : 12})
plt.xticks(k)

plt.show()
```
